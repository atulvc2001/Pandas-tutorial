# Making changes to the data

```python
# df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
# df = df.drop(columns=['Total']) # to remove a column

df['Total'] = df.iloc[:,4:10].sum(axis=1) # 1 means horizontally, 0 means vertically
# the above way is way more efficient to add cols

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

# Saving Data

```python
df.to_csv('modified.csv', index=False)
# to_excel
# to_csv('modified.txt', sep='\t')
```

# Filtering Data

# Aggregate Statistics (Group by)

```python
numcol = df.select_dtypes(include=['number']).columns
result = df.groupby(['Type 1'])[numcol].mean().sort_values('Attack', ascending=False)
result
```

```python
newres = df.groupby(['Type 1'])[numcol]
newres.count()
```

```python
df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

# Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
```
